```python
import pandas as pd
import numpy as np
```

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
```

```python
cars = pd.read_csv("imports-85.data", names = cols)
```

# Viewing data

```python
cars.head()
```

```python
#Use only the numeric columns for the model
```

```python
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
num_cars = cars[continuous_values_cols] #only taking numeric columns
```

```python
#view data again to make sure
num_cars.head(10)
```

```python
#Check for type of each column
num_cars.info()
```

# Clean Data

```python
#The "?" is a string so replace that
num_cars = num_cars.replace("?", np.nan)
#Check the data
num_cars.head()
```

```python
#Check info to see which columns to save
num_cars.info()
```

```python
num_cars = num_cars.astype('float')
num_cars.isnull().sum() #checking for number of missing entries
```

```python
#remove all rows missing price
#we can't use a data point if we don't have a price
num_cars = num_cars.dropna(subset=['price'])
num_cars.isnull().sum()
```

```python
#Replace all other missing value with the mean of their respected column
num_cars = num_cars.fillna(num_cars.mean())
#confirm for missing values
num_cars.isnull().sum() #should all be 0
```

```python
#Normalize all the data to avoid incorrect clasification
price = num_cars['price']
num_cars = (num_cars - num_cars.min())/(num_cars.max() - num_cars.min())
num_cars['price'] = price #put back in the original price col
```

# Univariate model

```python
#univariate first then extend to multivariate
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
def knn_train_test(inp_col, predict_col, df):
    np.random.seed(1)
    
    shuffle_ind = np.random.permutation(df.index)
    rand_df = df.reindex(shuffle_ind)
    
    end_at = int(len(rand_df)/2)
    
    train_df = rand_df.iloc[0:end_at]
    test_df = rand_df.iloc[end_at:]
    
    k_vals = range(1,10)
    k_rmses = {}
    
    for k in k_vals:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[inp_col]], train_df[predict_col])
    
        predicted_price = knn.predict(test_df[[inp_col]])
    
        mse = mean_squared_error(test_df[predict_col], predicted_price)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses

k_rmse_results = {}
train_cols = num_cars.columns.drop('price')

for col in train_cols:
    rmse_val = knn_train_test(col, 'price', num_cars)
    k_rmse_results[col] = rmse_val
    
k_rmse_results
```

```python
#now vizualize which K value gives a lower rmse
import matplotlib.pyplot as plt
```

```python
for k,v in k_rmse_results.items():
    x = list(v.keys())
    y = list(v.values())
    plt.plot(x,y, label="{}".format(k))
    

plt.xlabel('K value')
plt.ylabel('RMSE')
plt.legend()
```

# Multivariate model

```python
input_avg_rmse = {}
for k,v in k_rmse_results.items():
    avg_rmse = np.mean(list(v.values()))
    input_avg_rmse[k] = avg_rmse
series_avg_rmse = pd.Series(input_avg_rmse)
sorted_series = series_avg_rmse.sort_values()
print(sorted_series)
sorted_inp = sorted_series.index
print(sorted_inp)
```

```python
def knn_train_test(inp_cols, predict_col, df):
    np.random.seed(1)
    
    shuffle_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffle_index)
    
    end_row = int(len(rand_df)/2)
    
    train_df = rand_df.iloc[0:end_row]
    test_df = rand_df.iloc[end_row:]
    
    k_vals = [5]
    k_rmses = {}
    
    for k in k_vals:
        
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[inp_cols], train_df[predict_col])
        
        predictions = knn.predict(test_df[inp_cols])
        
        mse = mean_squared_error(test_df[predict_col], predictions)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses
```

# Feature selection

```python
#select good features
#The sorted series show features with the lowest rmse to the highest
#choose the lower ones to reduce redundant predictors
k_rmse_results = {}
for feat_ind in range(2,9):
    k_rmse_results[f"{feat_ind} best features"] = knn_train_test(sorted_inp[:feat_ind], 'price', num_cars)
    
k_rmse_results
```

# Hyperparameter Tuning

```python
#see which k-value produces the lowest rmse
def knn_train_test(inp_cols, predict_col, df):
    np.random.seed(1)
    
    shuffle_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffle_index)
    
    end_row = int(len(rand_df)/2)
    
    train_df = rand_df.iloc[0:end_row]
    test_df = rand_df.iloc[end_row:]
    
    k_vals = range(1,25)
    k_rmses = {}
    
    for k in k_vals:
        
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[inp_cols], train_df[predict_col])
        
        predictions = knn.predict(test_df[inp_cols])
        
        mse = mean_squared_error(test_df[predict_col], predictions)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses
```

```python
k_rmse_results = {}
for feat_ind in range(2,9):
    k_rmse_results[f"{feat_ind} best features"] = knn_train_test(sorted_inp[:feat_ind], 'price', num_cars)
    
k_rmse_results
```

```python
for k,v in k_rmse_results.items():
    x = list(v.keys())
    y = list(v.values())  
    plt.plot(x,y, label="{}".format(k))
    
plt.xlabel('k value')
plt.ylabel('RMSE')
plt.legend()
```

It seems that 2 features with k = 2 yields the lowest rmse value. So the engine-size and the horsepower seem like the best predictor for the price of a car.
